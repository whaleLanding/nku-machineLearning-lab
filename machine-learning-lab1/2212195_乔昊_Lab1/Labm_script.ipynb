```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

ex1data2=r"D:\machine-learning-lab1\ex1data2.txt"
data2=np.loadtxt(ex1data2,delimiter=',')
print(data2.shape,data2.ndim)
print(data2)
```

```python
x=data2[:,0:2]
y=data2[:,[2]]
print(x.shape,x.ndim)
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:,0],x[:,1],y)
```

```python
x[:,0] = ( x[:,0] - np.min(x[:,0]))/(np.max(x[:,0])-np.min(x[:,0]))
x[:,1] = ( x[:,1] - np.min(x[:,1]))/(np.max(x[:,1])-np.min(x[:,1]))
y = ( y - np.min(y))/(np.max(y)-np.min(y))
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],y)
```

```python
m=data2.shape[0] # data2.shaoe=（47,3）
x0=np.ones((m,1))
x1=data2[:,[0]]
x2=data2[:,[1]]
x=np.hstack((x0,x1,x2))
theta=np.zeros((3,1))
print(x)
print(theta)
```

```python
def computeCost(X,y,theta):
    h_theta_x=np.dot(X,theta)
    error=h_theta_x-y
    J=np.sum(error**2)/(2*m)
    return J
print(computeCost(x,y,theta))
```

```python
alpha = 0.01   
iterations = 30000   
cost=[]  
def gradientDescent(X,y,theta,alpha,iterations):
    
    for i in range(iterations):
         error = np.dot(X,theta) - y
         J = np.sum(error ** 2) / (2*m)
         grad=np.dot(error.T,X)/m
         theta = theta.T - grad * alpha
         theta=theta.T
         cost.append(J)
    return theta,cost
theta_star, cost = gradientDescent(x, y, theta, 0.01, 30000)
print(theta_star)
print(cost)
```

```python
# 优化求导法
T=np.dot(x.T,x)
x_reverse=np.linalg.inv(T)
var=np.dot(x_reverse,x.T)
theta=np.dot(var,y)
print(theta)
```
